```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from mplsoccer.pitch import Pitch, VerticalPitch
```

```python
#read data
passmap = pd.read_csv(r'C:\Users\alexa\OneDrive - Universidade do Minho\Documentos\BalizasPequenas\Tutoriais\PassMap\messibetis.csv')
passmap
```

```python
passmap['endX'] = passmap['endX']*1.2
passmap['endY'] = passmap['endY']*.8
passmap['x'] = passmap['x']*1.2
passmap['y'] = passmap['y']*.8
```

```python
passmap
```

```python
pitch = Pitch(pitch_type='statsbomb', goal_type='box', pitch_color='#aabb97', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#aabb97')



for x in range(len(passmap['x'])):
    if passmap['outcome'][x] == 'Successful':
        pitch.arrows(passmap['x'][x],passmap['y'][x],passmap['endX'][x],passmap['endY'][x],width=2,
             headwidth=10, headlength=10, color='green', ax=ax, label='completed passes')
    if passmap['outcome'][x] == 'Unsuccessful':
        pitch.arrows(passmap['x'][x],passmap['y'][x],passmap['endX'][x],passmap['endY'][x],width=2,
             headwidth=10, headlength=10,
             color='#ba4f45', ax=ax, label='other passes')
        
plt.title('Mapa de passes', size=20)
```

```python
#read data
ladoataques = pd.read_excel(r'C:\Users\alexa\OneDrive - Universidade do Minho\Documentos\BalizasPequenas\Tutoriais\LadoAtaqueTeste.xlsx')
ladoataques
```

```python
pitch = Pitch(pitch_type='statsbomb', goal_type='box', pitch_color='#aabb97', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(7, 10), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#aabb97')
text_color = 'green'
right_atacks = len(ladoataques)
central_atacks = len(ladoataques)
left_atacks = len(ladoataques)

for x in range(len(ladoataques['Event'])):
    if ladoataques['Event'][x] == 'Direita':
        pitch.arrows(20,10,100,10,width=2,
             headwidth=10, headlength=10, color='green', ax=ax, label='Direita')
        right_atacks += 1
    if ladoataques['Event'][x] == 'Esquerda':
        pitch.arrows(20,40,100,40,width=2,
             headwidth=10, headlength=10, color='green', ax=ax, label='Esquerda')
        left_atacks += 1
    if ladoataques['Event'][x] == 'Centro':
        pitch.arrows(20,70,100,70,width=2,
             headwidth=10, headlength=10, color='green', ax=ax, label='Centro')
        central_atacks += 1
        


fig_text(s=f'Ataques Pela Direita: {right_atacks}',
        x=.20, y =.38, fontsize=20,color=text_color)
fig_text(s=f'Ataques Pelo Centro: {central_atacks}',
        x=.20, y =.55, fontsize=20,color=text_color)
fig_text(s=f'Ataques Pela Esquerda: {left_atacks}',
        x=.20, y =.70, fontsize=20,color=text_color)

plt.title('Lado dos Ataques', size=20)
```
