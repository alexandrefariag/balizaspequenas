```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from mplsoccer import Pitch, VerticalPitch
import seaborn as sns
```

```python
df = pd.read_csv('../pass.csv')
df
```

```python
df['X'] = df['X']*1.2
df['Y'] = df['Y']*.8
df['X2'] = df['X2']*1.2
df['Y2'] = df['Y2']*.8
```

```python
df
```

```python
pitch = Pitch(pitch_type='statsbomb', goal_type='box', pitch_color='#aabb97', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#aabb97')



for x in range(len(df['X'])):
    if df['Event'][x] == 'Sucessful':
        pitch.arrows(df['X'][x],df['Y'][x],df['X2'][x],df['Y2'][x],width=2,
             headwidth=10, headlength=10, color='green', ax=ax, label='completed passes')
    if df['Event'][x] == 'Unsucessful':
        pitch.arrows(df['X'][x],df['Y'][x],df['X2'][x],df['Y2'][x],width=2,
             headwidth=10, headlength=10,
             color='#ba4f45', ax=ax, label='other passes')
        
plt.title('Pass Map', size=20)
```
